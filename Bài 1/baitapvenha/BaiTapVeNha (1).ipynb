```python
#baitapve nha
#bai1
import re

def nhapandxuat():
  A = list(map(int, input("Nhập Mảng A ").split()))
  B = list(map(int, input("Nhập Mảng B ").split()))
  c =[]
  i = 0
  n = len(A)
  m = len(B)
  for i in range(min(n, m)):
    c.append(min(A[i], B[i]))
  if n > m:
      c.extend(A[m:])
  else:
      c.extend(B[n:])
  print(f"Mảng Trước Khi Trộn \n Mảng A : {A} \n Mảng B : {B}")
  print("Mảng kết quả:",c)
nhapandxuat()
```

```python
#bai2
def nhapmatranthuong():
    m = int(input("Nhập số dòng: "))
    n = int(input("Nhập số cột: "))
    a = []
    print(f"Nhập ma trận {m}x{n}:")
    for i in range(m):
        while True:
            try:
                dong = list(map(int, input(f"Dòng {i+1}: ").split()))
                if len(dong) == n:
                    a.append(dong)
                    break
                else:
                    print(f"Lỗi: Phải nhập đúng {n} số.")
            except ValueError:
                print("Lỗi: Nhập các số nguyên cách nhau bởi dấu cách.")
    return a, m, n

def xuatmatran(a):
    for row in a:
        print("\t".join(map(str,row)))

def doiamthanhduong(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a

def thaysochan(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

def mangchan(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 != 0:
                return "Mảng không toàn chẵn"
    return "Mảng toàn chẵn"

def nhapmatranvuong():
    n = int(input("Nhập kích thước ma trận vuông: "))
    a = []
    print(f"Nhập ma trận vuông {n}x{n}:")
    for i in range(n):
        while True:
            try:
                dong = list(map(int, input(f"Dòng {i+1}: ").split()))
                if len(dong) == n:
                    a.append(dong)
                    break
                else:
                    print(f"Lỗi: Dòng {i+1} phải có {n} số.")
            except ValueError:
                print("Lỗi: Nhập các số nguyên cách nhau bởi dấu cách.")
    return a, n

def ktradoixung(a, n):
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return "Ma trận không đối xứng"
    return "Ma trận đối xứng"

def duongcheochinhtangdan(a):
    cheo = [a[i][i] for i in range(len(a))]
    if all(cheo[k] <= cheo[k+1] for k in range(len(cheo)-1)):
        return "Đường chéo chính đã tăng dần"
    else:
        return "Đường chéo chính không tăng dần"

def laytamgiac(a):
    n = len(a)
    tamgiac=[]
    for i in range(n):
        for j in range(n):
            if i + j >= n-1:
                tamgiac.append(a[i][j])
    return tamgiac

def laydgcheophu(a):
    n = len(a)
    cheophu=[]
    for i in range(n):
        for j in range(n):
            if i + j == n-1:
                cheophu.append(a[i][j])
    return cheophu

def cheophugiamdan(a):
    cheo = laydgcheophu(a)
    cheo.sort(reverse=True)
    return cheo

def main():
    while True:
        print("\n------ MENU CHÍNH ------")
        print("1. Thao tác trên ma trận thường")
        print("2. Thao tác trên ma trận vuông cấp n")
        print("0. Thoát")
        chon = input("Chọn chức năng: ")

        if chon == '1':
            a, m, n = nhapmatranthuong()
            while True:
                print("\n--- Menu Ma Trận Thường ---")
                print("1. Đổi âm thành dương")
                print("2. Thay số chẵn bằng x")
                print("3. Kiểm tra toàn chẵn")
                print("4. Xuất ma trận")
                print("0. Quay lại menu chính")
                lua = input("Chọn chức năng: ")
                if lua == '1':
                    doiamthanhduong(a)
                    print("Sau khi đổi âm -> dương:")
                    xuatmatran(a)
                elif lua == '2':
                    x = int(input("Nhập x: "))
                    thaysochan(a, x)
                    print("Sau khi thay số chẵn:")
                    xuatmatran(a)
                elif lua == '3':
                    print(mangchan(a))
                elif lua == '4':
                    print("Ma trận hiện tại:")
                    xuatmatran(a)
                elif lua == '0':
                    break
                else:
                    print("Lựa chọn không hợp lệ.")

        elif chon == '2':
            a, n = nhapmatranvuong()
            while True:
                print("\n--- Menu Ma Trận Vuông ---")
                print("1. Đường chéo chính tăng dần")
                print("2. Lấy tam giác dưới đường chéo phụ")
                print("3. Lấy đường chéo phụ")
                print("4. Đường chéo phụ giảm dần")
                print("5  . Xuất ma trận")
                print("0. Quay lại menu chính")
                lua = input("Chọn chức năng: ")
                if lua == '1':
                    print(duongcheochinhtangdan(a))
                elif lua == '2':
                    print("Tam giác dưới đường chéo phụ:", laytamgiac(a))
                elif lua == '3':
                    print("Đường chéo phụ:", laydgcheophu(a))
                elif lua == '4':
                    print("Đường chéo phụ giảm dần:", cheophugiamdan(a))
                elif lua == '5':
                    print("Ma trận hiện tại:")
                    xuatmatran(a)
                elif lua == '0':
                    break
                else:
                    print("Lựa chọn không hợp lệ.")

        elif chon == '0':
            print("Thoát chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ.")

main()
```
