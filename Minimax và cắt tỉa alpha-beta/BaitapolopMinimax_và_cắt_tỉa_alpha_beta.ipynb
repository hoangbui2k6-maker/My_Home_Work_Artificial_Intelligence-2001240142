```python
#bai 3
import math

def tao_ban_co(n):
    return [[" "]*n for _ in range(n)]

def kiem_tra_thang(ban_co, nguoi):
    n = len(ban_co)
    for i in range(n):
        if all(ban_co[i][j] == nguoi for j in range(n)):
            return True
        if all(ban_co[j][i] == nguoi for j in range(n)):
            return True
    if all(ban_co[i][i] == nguoi for i in range(n)):
        return True
    if all(ban_co[i][n-1-i] == nguoi for i in range(n)):
        return True
    return False

def la_hoa(ban_co):
    n = len(ban_co)
    return all(ban_co[i][j] != " " for i in range(n) for j in range(n))

def minimax(ban_co, la_max):
    if kiem_tra_thang(ban_co, "X"):
        return 1
    if kiem_tra_thang(ban_co, "O"):
        return -1
    if la_hoa(ban_co):
        return 0

    n = len(ban_co)

    if la_max:
        best = -math.inf
        for i in range(n):
            for j in range(n):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "X"
                    best = max(best, minimax(ban_co, False))
                    ban_co[i][j] = " "
        return best
    else:
        best = math.inf
        for i in range(n):
            for j in range(n):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "O"
                    best = min(best, minimax(ban_co, True))
                    ban_co[i][j] = " "
        return best

def nuoc_di_tot_nhat(ban_co):
    n = len(ban_co)
    best_val = -math.inf
    move = (-1, -1)

    for i in range(n):
        for j in range(n):
            if ban_co[i][j] == " ":
                ban_co[i][j] = "X"
                val = minimax(ban_co, False)
                ban_co[i][j] = " "
                if val > best_val:
                    best_val = val
                    move = (i, j)
    return move
```

```python
#bai 4
def alpha_beta(ban_co, alpha, beta, la_max):
    if kiem_tra_thang(ban_co, "X"):
        return 1
    if kiem_tra_thang(ban_co, "O"):
        return -1
    if la_hoa(ban_co):
        return 0

    n = len(ban_co)

    if la_max:
        best = -math.inf
        for i in range(n):
            for j in range(n):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "X"
                    best = max(best, alpha_beta(ban_co, alpha, beta, False))
                    ban_co[i][j] = " "
                    alpha = max(alpha, best)
                    if beta <= alpha:
                        return best
        return best
    else:
        best = math.inf
        for i in range(n):
            for j in range(n):
                if ban_co[i][j] == " ":
                    ban_co[i][j] = "O"
                    best = min(best, alpha_beta(ban_co, alpha, beta, True))
                    ban_co[i][j] = " "
                    beta = min(beta, best)
                    if beta <= alpha:
                        return best
        return best

def nuoc_di_tot_nhat_ab(ban_co):
    n = len(ban_co)
    best_val = -math.inf
    move = (-1, -1)

    for i in range(n):
        for j in range(n):
            if ban_co[i][j] == " ":
                ban_co[i][j] = "X"
                val = alpha_beta(ban_co, -math.inf, math.inf, False)
                ban_co[i][j] = " "
                if val > best_val:
                    best_val = val
                    move = (i, j)
    return move
```
