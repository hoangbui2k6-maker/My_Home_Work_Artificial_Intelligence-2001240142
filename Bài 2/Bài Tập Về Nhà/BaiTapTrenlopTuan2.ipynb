```python
import copy
from heapq import heappush, heappop

n = 3
dong = [1, 0, -1, 0]
cot = [0, -1, 0, 1]

class hangdoi:
    def __init__(self):
        self.heap = []

    def push(self, tkhoa):
        heappush(self.heap, tkhoa)

    def pop(self):
        return heappop(self.heap)

    def ktra(self):
        return not self.heap


class nodes:
    def __init__(self, nutdau, matran, vitri_o_trong, giatri, lop):
        self.nutdau = nutdau
        self.matran = matran
        self.vitri_o_trong = vitri_o_trong
        self.giatri = giatri
        self.lop = lop

    # sửa __It__ -> __lt__
    def __lt__(self, nxt):
        return (self.giatri + self.lop) < (nxt.giatri + nxt.lop)


# đưa ra ngoài class để gọi dễ
def heuristic(matran, vitri_cuoicung):
    dem = 0
    for i in range(n):
        for j in range(n):
            if matran[i][j] != 0 and matran[i][j] != vitri_cuoicung[i][j]:
                dem += 1
    return dem


def newnodes(matran, vitri_o_trong, otrong_moi, lop, nutcha, vitri_cuoicung):
    matranmoi = copy.deepcopy(matran)

    a1, b1 = vitri_o_trong
    a2, b2 = otrong_moi

    # hoán đổi
    matranmoi[a1][b1], matranmoi[a2][b2] = matranmoi[a2][b2], matranmoi[a1][b1]

    giatri = heuristic(matranmoi, vitri_cuoicung)

    node_moi = nodes(nutcha, matranmoi, otrong_moi, giatri, lop)
    return node_moi


def inmatran(matran):
    for i in range(n):
        for j in range(n):
            print("%d" % matran[i][j], end=" ")
        print()
    print()


def luugiatri(x, y):
    return x >= 0 and x < n and y >= 0 and y < n


def printtrang(root):
    if root is None:
        return
    printtrang(root.nutdau)
    inmatran(root.matran)


def giaiphap(initial, vitri_o_trong, vitri_cuoicung):
    pq = hangdoi()

    giatri = heuristic(initial, vitri_cuoicung)
    root = nodes(None, initial, vitri_o_trong, giatri, 0)

    pq.push(root)

    while not pq.ktra():
        minimum = pq.pop()

        if minimum.giatri == 0:
            print(" Solution:\n")
            printtrang(minimum)
            return

        for i in range(4):
            vitriomoi = (
                minimum.vitri_o_trong[0] + dong[i],
                minimum.vitri_o_trong[1] + cot[i]
            )

            if luugiatri(vitriomoi[0], vitriomoi[1]):
                nodecon = newnodes(
                    minimum.matran,
                    minimum.vitri_o_trong,
                    vitriomoi,
                    minimum.lop + 1,
                    minimum,
                    vitri_cuoicung
                )
                pq.push(nodecon)


# TEST
initial = [
    [1, 2, 3],
    [5, 6, 0],
    [7, 8, 4]
]

vitri_cuoicung = [
    [1, 2, 3],
    [4, 5, 6],
    [0, 8, 7]
]

vitri_o_trong = (1, 2)
giaiphap(initial, vitri_o_trong, vitri_cuoicung)
```

```python
#bài tập về nhà
#Bài 1. Cài đặt thuật giải AKT vào bài toán 15 puzzle(n=4).
#Bài 2. Cài đặt thuật giải A* tìm đường đi giữa 2 đỉnh bất kỳ trong đồ thị.
#bài 1
import heapq

class O_So:
    def __init__(self, trang_thai, muc_tieu, kich_thuoc_canh):
        self.trang_thai = trang_thai
        self.kich_thuoc_canh = kich_thuoc_canh
        self.muc_tieu = muc_tieu
        self.g = 0
        self.h = self.tinh_khoang_cach_manhattan()
        self.f = self.g + self.h
        self.cha = None

    def tinh_khoang_cach_manhattan(self):
        tong_khoang_cach = 0
        for i in range(len(self.trang_thai)):
            gia_tri = self.trang_thai[i]
            if gia_tri != 0:
                dong_hien_tai, cot_hien_tai = divmod(i, self.kich_thuoc_canh)

                chi_so_muc_tieu = self.muc_tieu.index(gia_tri)
                dong_muc_tieu, cot_muc_tieu = divmod(chi_so_muc_tieu, self.kich_thuoc_canh)

                tong_khoang_cach += abs(dong_hien_tai - dong_muc_tieu) + abs(cot_hien_tai - cot_muc_tieu)
        return tong_khoang_cach

    def __lt__(self, khac):
        return self.f < khac.f
def lay_cac_trang_thai_ke(nut_hien_tai):
    danh_sach_ke = []
    vi_tri_trong = nut_hien_tai.trang_thai.index(0)
    dong, cot = divmod(vi_tri_trong, nut_hien_tai.kich_thuoc_canh)

    huong_di = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for dr, dc in huong_di:
        dong_moi, cot_moi = dong + dr, cot + dc
        if 0 <= dong_moi < nut_hien_tai.kich_thuoc_canh and 0 <= cot_moi < nut_hien_tai.kich_thuoc_canh:
            mang_moi = list(nut_hien_tai.trang_thai)
            vi_tri_moi = dong_moi * nut_hien_tai.kich_thuoc_canh + cot_moi

            mang_moi[vi_tri_trong], mang_moi[vi_tri_moi] = mang_moi[vi_tri_moi], mang_moi[vi_tri_trong]
            danh_sach_ke.append(O_So(tuple(mang_moi), nut_hien_tai.muc_tieu, nut_hien_tai.kich_thuoc_canh))
    return danh_sach_ke

def giai_bai_toan_o_so(mang_dau, mang_dich, n):
    nut_bat_dau = O_So(tuple(mang_dau), tuple(mang_dich), n)
    tap_cho = [nut_bat_dau]
    tap_dong = set()
    while tap_cho:
        nut_hien_tai = heapq.heappop(tap_cho)

        if nut_hien_tai.trang_thai == nut_bat_dau.muc_tieu:
            return "Tim thay loi giai!"

        tap_dong.add(nut_hien_tai.trang_thai)

        for ke in lay_cac_trang_thai_ke(nut_hien_tai):
            if ke.trang_thai in tap_dong:
                continue
            ke.g = nut_hien_tai.g + 1
            ke.f = ke.g + ke.h
            ke.cha = nut_hien_tai
            heapq.heappush(tap_cho, ke)
    return "Khong co loi giai"
```

```python
#bai2
import math
import heapq

def thuat_toan_a_star_do_thi(do_thi, toa_do, diem_dau, diem_cuoi):
    tap_cho = [(0, diem_dau)]
    nguon_goc = {}
    chi_phi_g = {nut: float('inf') for nut in do_thi}
    chi_phi_g[diem_dau] = 0

    while tap_cho:
        _, nut_hien_tai = heapq.heappop(tap_cho)

        if nut_hien_tai == diem_cuoi:
            return "Da den dich"

        for hang_xom, trong_so in do_thi[nut_hien_tai].items():
            g_tam_thoi = chi_phi_g[nut_hien_tai] + trong_so

            if g_tam_thoi < chi_phi_g[hang_xom]:
                nguon_goc[hang_xom] = nut_hien_tai
                chi_phi_g[hang_xom] = g_tam_thoi


                h = math.sqrt((toa_do[hang_xom][0] - toa_do[diem_cuoi][0])**2 +
                              (toa_do[hang_xom][1] - toa_do[diem_cuoi][1])**2)

                chi_phi_f = g_tam_thoi + h
                heapq.heappush(tap_cho, (chi_phi_f, hang_xom))
    return None
```
